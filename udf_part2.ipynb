```python
def to_lower(text):
    print("this is lower function")
    return text.lower()
```

```python
def to_upper(text):
    return text.upper()
```

```python
to_upper("abc")
```

```python
def wrapper(func, text):
    """ this is a wrapper function which takes a function and an argument which has to be sent to the function as argument """
    wrap = func(text)
    print(wrap)
```

```python
wrapper.__doc__
```

```python
to_lower("tHis is junk TEXT")
```

```python
wrapper(to_lower, "tHis is junk TEXT")
```

```python
wrapper(to_upper, "tHis is MORE junk TEXT")
```

- decorators

```python
def make_pretty(func):
    def inner():
        print("decorated")
        func()
    return inner
```

```python
# @make_pretty
def ordinary():
    print("oridinary")
```

```python
ordinary()
```

```python
@make_pretty
def ordinary():
    print("oridinary")
```

```python
ordinary()
```

```python
fn = make_pretty(ordinary)
fn()
```

```python
@make_pretty
def ordinary():
    print("oridinary")
```

```python
ordinary()
```

- let us use decorators to calculate time a function takes to operate

```python
import time
import math
```

```python
def calculate_time(func):
    def core_time_calculator(*args, **kwargs):
        start_time = time.time()
        func(*args, **kwargs)
        end_time = time.time()
        print(f"Time taken to execute is :{end_time-start_time} seconds")
    return core_time_calculator
```

```python
math.factorial(10)
```

```python
def factorial(num):
    time.sleep(2)
    print(math.factorial(num))
```

```python
factorial(5)
```

```python
@calculate_time
def factorial(num):
    time.sleep(2)
    print(math.factorial(num))
```

```python
factorial(10)
```

```python
@calculate_time
def add_three_nums_kw_all(x=0, y=0, z=0):
    return x+y+z
```

```python
add_three_nums_kw_all(5, 3, 8)
```
